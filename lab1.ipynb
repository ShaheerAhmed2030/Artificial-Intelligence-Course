#### TASK 4b: Now take a sample digit 3 or digit 7 and finds its distance from the ideal digit 3 and digit 7.

We cannot just take the difference between the ideal 3 and the sample digit and add them. Why? Because some of the pixel differences would be positive and some would be negative. Adding them would result in an image which is too dark or too light at different places and may have zero total difference from the ideal. This would be misleading!

To avoid this, there are two ways distances are taken.

- Take the mean of the *absolute value* of differences (absolute value is the function that replaces negative values with positive values). This is called the *mean absolute difference* or *L1 norm*
- Take the mean of the *square* of differences (which makes everything positive) and then take the *square root* (which undoes the squaring). This is called the *root mean squared error* (RMSE) or *L2 norm*.

```python
"""
Todo: Write a function for both L1, and L2 norm. Also find out the L1 and L2 norm of a digit from the ideal 3 and 7. 
"""
def MeanAbsoluteDistance(Set,Ideal):
    return np.abs(np.subtract(Set,Ideal)).mean()
    #Difference = np.abs(np.subtract(Set,Ideal))
    #return np.mean(Difference)

def RootMeanSquared(Set,Ideal):
    return np.sqrt(np.square(np.subtract(Set,Ideal)).mean())

print(MeanAbsoluteDistance(Set_3[0],Ideal3))
print(MeanAbsoluteDistance(Set_7[0],Ideal7))
print(RootMeanSquared(Set_3[0],Ideal3))
print(RootMeanSquared(Set_7[1],Ideal7))
```

#### Task 5: Find out how well did we do on all the validation set?
We can find out how well our algorithm performs by calculating the distance of each digit in our validation set with the ideal 3 and 7, and then by comparing if the distance is closer to the actual label or not. This tells us how good our algorithm is, and is usually known as *metric*. In practice, accuracy is one metric that we use.

We will calculate our metric i.e accuracy over the validation set.

#### Task 5a: Create a subsample of validation set containing only digit 3, and digit 7. Ignore if you have done it in Task 3.

#### Task 5b: Now using any of the distance function from task 4b, write down a function that tells you if a digit is 3 or 7.


```python
"""
Todo: Create a function that tells you if an image contains digit 3. You will need to use the distance function you have created in previous task.
"""
def ThreeORSevenUsingL1Norm(Img,Ideal3,Ideal7):
    Three = MeanAbsoluteDistance(Img,Ideal3)
    Seven = MeanAbsoluteDistance(Img,Ideal7)
    if Three < Seven:
        return 3
    else:
        return 7


def ThreeORSevenUsingL2Norm(Img,Ideal3,Ideal7):
    Three = RootMeanSquared(Img,Ideal3)
    Seven = RootMeanSquared(Img,Ideal7)
    if Three < Seven:
        return 3
    else:
        return 7
```

```python
"""
Todo: Calculate accuracy over the validation set.
"""
# Accuracy using L1 Norm
Count = 0
for i in range(len(valSet_3)):
    if ThreeORSevenUsingL1Norm(valSet_3[i],Ideal3,Ideal7) == y_valSet_3[i]:
        Count += 1
for i in range(len(valSet_7)):
    if ThreeORSevenUsingL1Norm(valSet_7[i],Ideal3,Ideal7) == y_valSet_7[i]:
        Count += 1
Total = len(valSet_3) + len(valSet_7)
print("Accuracy using L1 Norm:",(Count/Total)*100)

# Accuracy using L2 Norm
Count = 0
for i in range(len(valSet_3)):
    if ThreeORSevenUsingL2Norm(valSet_3[i],Ideal3,Ideal7) == y_valSet_3[i]:
        Count += 1
for i in range(len(valSet_7)):
    if ThreeORSevenUsingL2Norm(valSet_7[i],Ideal3,Ideal7) == y_valSet_7[i]:
        Count += 1
Total = len(valSet_3) + len(valSet_7)
print("Accuracy using L2 Norm:",(Count/Total)*100)
```

## Classifying 3 or 7 using a Machine Learning Model

In Machine Learning, we have some initial values of parameters that are used to find out how good the model is performing currently on the task. This is usually done using a function that behaves as a metric for the model. Usually known as *loss function*. Lower the value of loss function, the better our model is at the task.

After finding our how good our model is currently behaving, using the value of the loss function, we change the values of our inital parameters to obtain a smaller loss function value. Each time we change the parameter values, we also find out the accuracy again to see if our machine learning model is improving i.e accuracy is improving. This process is repeated until we are satisfied with the accuracy value obtained. 



```python
import tensorflow as tf
```

```python
def get_fully_connected_network(intralayer_activation='sigmoid'):
    inputs = tf.keras.Input(shape=(28,28), name="digits")
    flatten = tf.keras.layers.Flatten()(inputs)
    x1 = tf.keras.layers.Dense(100, activation=intralayer_activation)(flatten)
    outputs  = tf.keras.layers.Dense(1, activation=intralayer_activation)(x1)
    model_fullyconnect = tf.keras.Model(inputs, outputs)
    return model_fullyconnect
model_fullyconnect = get_fully_connected_network()
```

```python
(x_train, y_train), (x_val, y_val) = tf.keras.datasets.mnist.load_data()
```

```python
import numpy as np
```

```python
indices_3 = np.where(y_train==2)[0]
indices_7 = np.where(y_train==6)[0]
dataset_train = np.concatenate((x_train[indices_3],x_train[indices_7]))


```

```python
labels_train = []
for i in range(indices_3.shape[0] + indices_7.shape[0]):
    if i < indices_3.shape[0]:
        labels_train.append(0)
    else:
        labels_train.append(1)
labels_train = np.array(labels_train)

indices_3 = np.where(y_val==2)[0]
indices_7 = np.where(y_val==6)[0]
dataset_val = np.concatenate((x_val[indices_3],x_val[indices_7]))


labels_val = []
for i in range(indices_3.shape[0] + indices_7.shape[0]):
    if i < indices_3.shape[0]:
        labels_val.append(0)
    else:
        labels_val.append(1)
labels_val = np.array(labels_val)
```

```python
model_fullyconnect.compile(optimizer="SGD", loss=tf.keras.losses.BinaryCrossentropy(), metrics=["accuracy"] )
```

```python
model_fullyconnect.fit(dataset_train, labels_train, 64,validation_data=(dataset_val,labels_val), epochs=60)
```
